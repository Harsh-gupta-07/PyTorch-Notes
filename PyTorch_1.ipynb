## Manipulating Tensors

```python
#add
tensor = torch.tensor([1,2,3])
tensor + 10
```

```python
#subtract
tensor - 1
```

```python
#multiply
tensor * 100
torch.mul(tensor, 10)
```

### Matrix Multiply

```python
#element wise
tensor = torch.tensor([1,2,3])
tensor * tensor
```

```python
#matrix
torch.matmul(tensor, tensor)
```

## Reshaping, stacking, sqeezing and unsqueezing tensors

```python
x = torch.arange(1.,11.)
x, x.shape
```

```python
x_re = x.reshape(2,5)
x_re, x_re.shape
```

```python
# shares memory with x
z = x.view(2,5)
z
```

###Stacking

```python
torch.stack([x,x,x,x], dim=0)
```

```python
torch.stack([x,x,x,x], dim=1)
```

### Squeezing

```python
y = x.reshape(1,10)
y
```

```python
# remove single dimensions (dim with one element)
y_sq = y.squeeze_()
y_sq
```

```python
# adds a single dimension if shape [10] on dim=0 => [1,10] or dim=1 => [10,1]
y_sq.unsqueeze(dim=0)
```

### Permute

```python
x_original = torch.rand(size=(224,224,5))
```

```python
#swap dimensions based on index
x_pre = x_original.permute(2,0,1)
```

```python
x_original.shape, x_pre.shape
```
